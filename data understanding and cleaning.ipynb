### dropping unwanted column

```python
df1 = df.drop(columns=['ISRC','TIDAL Popularity'],axis=1)
df1.head()
```

### Filling Missing Values 

```python
df1['Artist'] = df1['Artist'].fillna('Unknown')
```

```python
columns_to_fill = [ 'Spotify Streams', 'Spotify Playlist Count',
       'Spotify Playlist Reach', 'Spotify Popularity', 'YouTube Views',
       'YouTube Likes', 'TikTok Posts', 'TikTok Likes', 'TikTok Views',
       'YouTube Playlist Reach', 'Apple Music Playlist Count', 'AirPlay Spins',
       'SiriusXM Spins', 'Deezer Playlist Count', 'Deezer Playlist Reach',
       'Amazon Playlist Count', 'Pandora Streams', 'Pandora Track Stations',
       'Soundcloud Streams', 'Shazam Counts']
df1[columns_to_fill] = df1[columns_to_fill].fillna(0)
```

### Data after filling missing values

```python
msno.bar(df1)
```

### Removing commas and changing dtype of column into float

```python
columns_to_convert = ['All Time Rank','Spotify Streams','Spotify Playlist Count',
                     'Spotify Playlist Reach','YouTube Views','YouTube Likes',
                     'TikTok Posts','TikTok Likes','TikTok Views','YouTube Playlist Reach','AirPlay Spins','SiriusXM Spins',
                    'Deezer Playlist Reach','Pandora Streams','Pandora Track Stations','Soundcloud Streams','Shazam Counts']
```

```python
for column in columns_to_convert:
    df1[column] = pd.to_numeric(df1[column].str.replace(',','').astype(float),errors='coerce')
```

```python
df1.info()
```

### Converting date column  to datetime format

### Removing whitespaces if any

```python
df1['Release Date'] = df1['Release Date'].str.strip()
```

```python
df1['Release Date'] = pd.to_datetime(df1['Release Date'], errors='coerce')
```

```python
df1['Release Date'].head()
```

### Creating New Column

```python
df1['Year'] = df1['Release Date'].dt.year
```

```python
df1.head()
```

### Just taking look at corelated columns

```python
correlation_matrix = df1.corr()
plt.figure(figsize=(20,18))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation matrix")
```
