```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Most Streamed Spotify Songs 2024(Updated).csv')
df.head()
```

### Average Track Score for each Year

```python
avg_track_score_per_year = df.groupby("Year")['Track Score'].mean()
# avg_track_score_per_year
plt.figure(figsize=(10, 8))
plt.plot(avg_track_score_per_year,marker='o')
plt.title('Change in Track Score Throughout the Years')
plt.xlabel('Year')
plt.ylabel('Average Track Score')
plt.grid(True)
```

### Best Track Score for each Year

```python
best_track_score_per_year = df.groupby("Year")['Track Score'].max()
# avg_track_score_per_year
plt.figure(figsize=(10, 8))
plt.plot(best_track_score_per_year,marker='o')
plt.title('Change in Track Score Throughout the Years')
plt.xlabel('Year')
plt.ylabel('Average Track Score')
plt.grid(True)
```

### Worst Track Score for each Year

```python
worst_track_score_per_year = df.groupby("Year")['Track Score'].min()
# worst_track_score_per_year
plt.figure(figsize=(10, 8))
plt.plot(worst_track_score_per_year,marker='o')
plt.title('Change in Track Score Throughout the Years')
plt.xlabel('Year')
plt.ylabel('Worst Track Score')
plt.grid(True)
```

### Number of tracks each year

```python
no_of_tracks_each_year = df.groupby('Year')['Track'].count()
# no_of_tracks_each_year

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# line chart
no_of_tracks_each_year.plot(kind="line", ax=axes[0] , color="blue",marker='o')
axes[0].set_title('Number of Tracks every year')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Number of Tracks')
axes[0].grid(True)

# bar chart
no_of_tracks_each_year.plot(kind="bar", ax=axes[1] ,color="orange")
axes[1].set_title('Number of Tracks every year')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Number of Tracks')
axes[1].grid(True)
```

### Intrest in Explicit Track

```python
Explicit_track_score = df.groupby('Explicit Track')['Track Score'].mean()

plt.figure(figsize = (12,8))
Explicit_track_score.plot(kind='bar',color="skyblue")
plt.title("Track score depending on explicitness")
plt.xlabel("Explicit track")
plt.ylabel("Test score")
```
